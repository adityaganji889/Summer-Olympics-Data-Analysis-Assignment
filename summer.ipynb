### 1. In how many cities Summer Olympics is held so far?

```python
sum_city=df['City'].unique()
print("Different cities list in which Summer Olympics is held so far:")
print(sum_city)
print("\n")
print("No.of cities in which Summer Olympics is held so far:")
print(len(sum_city))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_1=df[df['Medal']=='Gold']
data = []
for Sport in df_1['Sport'].unique():
    data.append([Sport , len(df_1[df_1['Sport']  == Sport])])
data_frame=pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False)
dat=data_frame.head()
print("Top 5 sports with most number of Gold Medals:")
print(dat)
print("\n")
print("Bar graph for the top 5 sports having most number of Gold Medals:")
print(dat.plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5),color='yellow'))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []
for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']==Sport])])
data_frame=pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False)
dat=data_frame.head()
print("Top 5 sports with most number of medals so far:")
print(dat)
print("\n")
print("Bar graph for the top 5 sports having most number of medals so far:")
print(dat.plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5)))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []
for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']==Athlete])])
data_frame=pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False)
dat=data_frame.head()
print("Top 5 players who have won most number of medals:")
print(dat)
print("\n")
print("Bar graph for the top 5 players who have won most number of medals:")
print(dat.plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (5,5),color='orange'))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []
for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']==Event])])

data_frame=pd.DataFrame(data,columns = ['Event','Athlete']).sort_values(by='Athlete', ascending=False)
dat=data_frame.head()
print("Top 5 events which are popular in terms on number of players:")
print(dat)
print("\n")
print("Bar graph for the top 5 events which are popular in terms on number of players:")
print(dat.plot(x = 'Event', y = 'Athlete', kind = 'bar', figsize = (5,5),color='lightgreen'))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df_1=df[df['Medal']=='Gold']
df_2=df_1[df_1['Gender']=='Women']
data = []
for Sport in df_2['Sport'].unique():
    data.append([Sport , len(df_2[df_2['Sport']  == Sport])])
data_frame=pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False)
dat=data_frame.head()
print("Top 5 sports having most female Gold Medalists:")
print(dat)
print("\n")
print("Bar graph for the top 5 sports having most female Gold Medalists:")
print(dat.plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5),color='brown'))
```
